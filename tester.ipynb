```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns   # if you don't have this: pip install seaborn

from scipy.stats import skew, kurtosis
from numpy import mean

def diagnose_distributions(df):
    report = []
    series_list = df['series_name'].unique()
    
    print(f"{'SERIES NAME':<40} | {'MEAN':<10} | {'SKEW':<10} | {'KURTOSIS':<10} | {'STATUS'}")
    print("-" * 85)
    
    for s in series_list:
        # Get data, drop NaNs to avoid calculation errors
        data = df[df['series_name'] == s]['value'].dropna()
        
        if len(data) < 10:
            continue
            
        m_val = mean(data)    
        s_val = skew(data)
        k_val = kurtosis(data)  # Fisher kurtosis (normal = 0)
        
        # Heuristic for "Weird"
        if abs(s_val) > 2.0 or abs(k_val) > 7.0:
            status = "ðŸš¨ EXTREME"
        elif abs(s_val) > 1.0 or abs(k_val) > 3.0:
            status = "âš ï¸  SKEWED"
        else:
            status = "âœ… OK"
            
        print(f"{s:<40} | {m_val:6.2f}   |  {s_val:6.2f}     | {k_val:6.2f}     | {status}")


def correlation_analysis(
    df,
    date_col="date",
    series_col="series_name",
    value_col="value",
    top_n=150
):
    """
    Build a wide matrix of series, plot correlation heatmap,
    and print top-N most correlated distinct pairs.
    """
    # 1) Pivot to wide: rows = dates, columns = series, values = value
    wide = df.pivot(index=date_col, columns=series_col, values=value_col)
    
    # 2) Sort series alphabetically
    wide = wide[sorted(wide.columns)]
    
    # 3) Correlation matrix
    corr = wide.corr()
    
    # --- Heatmap visualization ---
    import matplotlib.pyplot as plt
    import seaborn as sns
    
    plt.figure(figsize=(14, 12))
    sns.heatmap(
        corr,
        cmap="coolwarm",
        center=0,
        square=True,
        xticklabels=True,
        yticklabels=True
    )
    plt.title("Correlation Matrix of Series")
    plt.tight_layout()
    plt.show()
    
    # --- Top-N pairwise correlations ---
    import numpy as np
    
    # Keep only upper triangle (excluding diagonal) to avoid duplicates
    mask = np.triu(np.ones_like(corr, dtype=bool), k=1)
    corr_upper = corr.where(mask)

    # Stack to long format: (series_1, series_2, corr)
    pairs = (
        corr_upper
        .stack()                                # -> Series with MultiIndex
        .rename_axis(["series_1", "series_2"])  # rename index levels here
        .reset_index(name="correlation")        # correlation column
    )
    
    # Sort by absolute correlation strength
    pairs["abs_corr"] = pairs["correlation"].abs()
    top_pairs = pairs.sort_values("abs_corr", ascending=False).head(top_n)
    
    # Nice printing
    print(f"\nTop {top_n} most correlated series pairs (by |corr|):\n")
    print(f"{'SERIES 1':<35} | {'SERIES 2':<35} | {'CORR':>7}")
    print("-" * 85)
    for _, row in top_pairs.iterrows():
        print(f"{row['series_1']:<35} | {row['series_2']:<35} | {row['correlation']:7.3f}")
    
    return corr, top_pairs


# Example usage on your verified master dataframe:
# First run your distribution diagnostics:
diagnose_distributions(df_long_clean)

# Then run correlation analysis:
corr_matrix, top_50_pairs = correlation_analysis(df_long_clean, date_col="date")
```

```python
import pandas as pd
import numpy as np
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant

def calculate_vif_from_long_format(df):

    df_wide = df.pivot_table(index='date', columns='series_name', values='value', aggfunc='mean')

    df_wide = df_wide.ffill().dropna()
    
    if df_wide.empty:
        return "Error: DataFrame is empty after handling missing values. Check your date alignment."
    X = add_constant(df_wide)
    
    vif_data = pd.DataFrame()
    vif_data["feature"] = X.columns
    

    vif_data["VIF"] = [variance_inflation_factor(X.values, i) 
                       for i in range(X.shape[1])]
    
    vif_data = vif_data[vif_data["feature"] != "const"]
    return vif_data.sort_values(by="VIF", ascending=False)

# Usage Example:
vif_report = calculate_vif_from_long_format(df_long_clean)
print(vif_report)
```

```python
def iterative_vif_selection(df, threshold=10.0):
    """
    Iteratively removes the feature with the highest VIF until all VIFs are below the threshold.
    """
    # 1. Pivot to wide format first (using your snapshot logic)
    df_wide = df.pivot_table(index='date', columns='series_name', values='value', aggfunc='mean').ffill().dropna()
    
    # 2. Iteration Loop
    features_to_drop = []
    
    while True:
        # Define current features (excluding dropped ones)
        current_cols = df_wide.columns.difference(features_to_drop)
        X = add_constant(df_wide[current_cols])
        
        # Calculate VIFs
        vif_scores = pd.Series(
            [variance_inflation_factor(X.values, i) for i in range(X.shape[1])],
            index=X.columns
        )
        # Drop constant
        vif_scores = vif_scores.drop('const')
        
        # Check highest VIF
        max_vif = vif_scores.max()
        worst_feature = vif_scores.idxmax()
        
        if max_vif > threshold:
            print(f"Dropping '{worst_feature}' with VIF: {max_vif:.1f}")
            features_to_drop.append(worst_feature)
        else:
            print(f"Optimization Complete! All VIFs below {threshold}.")
            break
            
    print("\nFinal Feature Set:", list(df_wide.columns.difference(features_to_drop)))
    return df_wide.drop(columns=features_to_drop)

# Run it
lean_df = iterative_vif_selection(df_long_clean, threshold=10)
```

```python
def drop_high_vif_features(df, series_col='series_name'):
    """
    Removes specific features identified as having high VIF (Variance Inflation Factor).
    Assumes a long-format dataframe where variable names are in 'series_col'.
    """
    
    # The specific list of variables to drop based on your VIF analysis
    # We include ADP_forecast to be safe, in case it was re-introduced or missed.
    to_drop = [
        'ADP_previous',
        'IURSA_monthly_avg',
        'ICSA_monthly_avg',
        'JOLTS_Hires',
        'JOLTS_Openings',
        'ADP_forecast' 
    ]
    
    # Create a copy to avoid SettingWithCopy warnings on the original df
    df_clean = df.copy()
    
    # Filter: Keep rows where the series_name is NOT in the 'to_drop' list
    df_clean = df_clean[~df_clean[series_col].isin(to_drop)]
    
    # Verification print
    print(f"Original Row Count: {len(df)}")
    print(f"Cleaned Row Count:  {len(df_clean)}")
    print(f"Dropped Series:     {to_drop}")
    
    return df_clean

# --- usage ---
df_long_final = drop_high_vif_features(df_long_clean)
```

```python
import pandas as pd

def find_nan_runs_in_series(s: pd.Series):
    """
    Find all contiguous NaN runs in a single pandas Series.

    Returns a list of dicts:
        [{"start": index_start, "end": index_end, "length": run_length}, ...]
    """
    mask = s.isna()
    if not mask.any():
        return []

    # Identify groups where the NaN/non-NaN status changes
    group_id = (mask != mask.shift()).cumsum()

    runs = []
    # We only care about groups where mask is True (NaN)
    nan_group_ids = group_id[mask].unique()

    for gid in nan_group_ids:
        this_run_mask = (group_id == gid) & mask
        idx = s.index[this_run_mask]

        start = idx[0]
        end = idx[-1]
        length = len(idx)

        runs.append({"start": start, "end": end, "length": length})

    return runs


def find_nan_runs_in_dataframe(df: pd.DataFrame) -> pd.DataFrame:
    """
    For each column in df, find all contiguous NaN runs.

    Returns a long DataFrame with columns:
        ['series', 'run_id', 'start', 'end', 'length']
    """
    records = []
    for col in df.columns:
        runs = find_nan_runs_in_series(df[col])
        for i, r in enumerate(runs, start=1):
            records.append({
                "series": col,
                "run_id": i,
                "start": r["start"],
                "end": r["end"],
                "length": r["length"],
            })

    if not records:
        return pd.DataFrame(columns=["series", "run_id", "start", "end", "length"])

    return pd.DataFrame(records)
```
